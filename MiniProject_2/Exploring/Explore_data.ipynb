##### <span style="color:#87CEEB">Finding dependent and independent variables</span>

After taking a look at the wine dataset it is clear that the variable that we should try to understand is the wine quality. Our analysis goal would in this case be to explore what effects or determines the quality of wine - what makes the wine higher or lower in quality based on the other characteristics of the wine that are measurable?

Based on this goal we can then say that the "quality" variable is a dependent variable while the other variables are independent. 
- The quality of the wine is dependent on other variables of the wine to determine if it's high or low - the outcome of the quality changes in response to the independent variables, therefore making it a dependent variable.
- The other variables of the wine could influence or explain the quality of wine - it acts as input to the outcome of the dependent variable and are not influenced by other variables, therefore making them independent.

```python
dependent = 'quality'
all_independent = [col for col in df_all.columns if col not in ['quality']]

print("All Independent variables:")
all_independent
```

#### Task 5: Transform the categorical data into numeric, applying appropriate encoding methods.

Looking at the wine dataset for all wines, we can tell from the info-method that the only variable of type object is the "wine_type" variable. 
Since this is our only categorical data we can then encode it. 

The appropriate encoding method for this case would be to use "One-hot Encoding" just for the simple reason that there's no implied ranking order for the types of wine, compared to using "Label Encoding". Since our categorical data is of the nominal type, which means it's just purely descriptive, then they dont have an order and can't be compared.

To avoid multicollinearity we collect the wine type in one column "wine_type_white" and have either 0 or 1 to represent if it's white or red wine:
- 0: Means that it's false so the wine is not white, therefore it is red
- 1: Means that it's true so the wine is white

```python
df_all = pd.get_dummies(df_all, columns=['wine_type'], drop_first=True)
df_all['wine_type_white'] = df_all['wine_type_white'].astype(int)

df_all.sample(10)
```

##### <span style="color:#87CEEB">Are there any attributes, apart from the wine quality, which are highly correlated?</span>

```python
## Unstack, sort and filter out weak correlations

high_correlation_pairs = (
    correlation_matrix
    .abs()
    .unstack()
    .sort_values(ascending=False)
    .drop_duplicates()
)
```

```python
## Removes self-correlations, any rows where the first variable in the pair is quality and assuring only strong correlation above 0.5.

high_correlation_pairs = high_correlation_pairs[
    (high_correlation_pairs < 1.0) &
    (~high_correlation_pairs.index.get_level_values(0).isin(['quality'])) & ## ~ = (not)
    (~high_correlation_pairs.index.get_level_values(1).isin(['quality'])) &
    (high_correlation_pairs > 0.5)
]
```

```python
high_correlation_pairs = high_correlation_pairs.to_frame(name='Correlation (r)')
print("Highly correlated attribute pairs (r > 0.5):")
print(high_correlation_pairs)
```

We unstacked the matrix to get a list of all possible variable pairs and applied filters to:

* Remove self-correlations (r = 1.0)

* Exclude any pairs that involved the dependent variable quality

* Keep only strong correlations with r > 0.5 for meaningful relationships

##### <span style="color:#87CEEB">Scatter plot (Visually confirmation of highly correlations in the heatmap ) </span>

```python
## List of correlation between specific attribute pairs
pairs = [
    ('alcohol', 'quality'),
    ('total_sulfur_dioxide', 'free_sulfur_dioxide'),
    ('density', 'alcohol'),
    ('wine_type_white','total_sulfur_dioxide'),
    ('density', 'residual_sugar'),
    ('wine_type_white','volatile_acidity')
]
```

```python
## Visual representation of attribute pairs
for x, y in pairs:
    plt.figure(figsize=(6, 4))
    sns.regplot(data=df_all, x=x, y=y)
    plt.title(f"{y} vs {x}")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

The scatter plot  shows the relationship between two wine attributes. Every dot represents one wine sample, placed according to its value for the X and Y variables.

The blue regression line shows the overall trend in the data: 
* If the line slopes upward, it means there's a positive correlation: as X increases, Y tends to increase. 
* If the line slopes downward, there’s a negative correlation: as X increases, Y tends to decrease. 
* If the line is flat or the points are widely scattered, there's likely no strong correlation.

These scatter plots confirm different patterns:
* There's a clear upward trend between alcohol and quality, and a strong positive relationship between total_sulfur_dioxide and free_sulfur_dioxide. 
* In contrast, wine_type_white vs volatile_acidity shows a negative trend, suggesting that white wines tend to have lower acidity.
