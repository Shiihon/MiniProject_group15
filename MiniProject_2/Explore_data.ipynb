##### <span style="color:#87CEEB">Finding dependent and independent variables</span>

After taking a look at the wine dataset it is clear that the variable that we should try to understand is the wine quality. Our analysis goal would in this case be to explore what effects or determines the quality of wine - what makes the wine higher or lower in quality based on the other characteristics of the wine that are measurable?

Based on this goal we can then say that the "quality" variable is a dependent variable while the other variables are independent. 
- The quality of the wine is dependent on other variables of the wine to determine if it's high or low - the outcome of the quality changes in response to the independent variables, therefore making it a dependent variable.
- The other variables of the wine could influence or explain the quality of wine - it acts as input to the outcome of the dependent variable and are not influenced by other variables, therefore making them independent.

```python
dependent = 'quality'
all_independent = [col for col in df_all.columns if col not in ['quality']]

print("All Independent variables:")
all_independent
```

#### Task 5: Transform the categorical data into numeric, applying appropriate encoding methods.

Looking at the wine dataset for all wines, we can tell from the info-method that the only variable of type object is the "wine_type" variable. 
Since this is our only categorical data we can then encode it. 

The appropriate encoding method for this case would be to use "One-hot Encoding" just for the simple reason that there's no implied ranking order for the types of wine, compared to using "Label Encoding". Since our categorical data is of the nominal type, which means it's just purely descriptive, then they dont have an order and can't be compared.

To avoid multicollinearity we collect the wine type in one column "wine_type_white" and have either 0 or 1 to represent if it's white or red wine:
- 0: Means that it's false so the wine is not white, therefore it is red
- 1: Means that it's true so the wine is white

```python
df_all = pd.get_dummies(df_all, columns=['wine_type'], drop_first=True)
df_all['wine_type_white'] = df_all['wine_type_white'].astype(int)

df_all.sample(10)
```

##### <span style="color:#87CEEB">Calculate the descriptive statistcs</span>

We start out by calculating the descriptive statistics to get a summary of each numeric column

```python
df_all.describe()
```

Based on the above output we can then break it down and explain the metrics:

- The "count" metric is the number of non-null entries found in our dataframe, which is equal across all columns since we have no missing values.

- The "mean" metric is the average value and gives us an idea of the central tendency.

- The "std" metric is the standard deviation, which tells us the spread of the data. A high std means that values may vary a lot from the mean, while a low std points to tightly grouped values.

- The "min" metric is the minimum value and is useful for spotting outliers or floor effects. 
- The "max" metric is the maximum value and is useful for spotting outliers or ceiling effects. 

- The "25%" metric represents the first quartile (Q1), which tells us that 25% of the data falls below this value.

- The "50%" metric represents the median which is the middle point of the data. It tells us that 50% of the data is below this point.

- The "75%" metric represents the third quartile (Q3), which tells us that 75% of the data falls below this value.

##### <span style="color:#87CEEB">Check the distribution of the values of the attributes</span>

###### <span style="color:#f1c40f">Histograms</span>

Histograms shows how the data is spread across interval and is useful for to see if the distribution is normal, skewed, uniform etc.

To check whether the distribution of the values of the attributes is normal we can easily determine that based on a bell-shaped histogram, while one with long tail to one side suggests skewness.

```python
# Get all numeric columns
numeric_cols = df_all.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    sns.histplot(df_all[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```
